```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

st = pd.read_csv("student-mat.csv", sep=',')
```

```python
st = st[["G1","G2","G3","age","Medu","Fedu","absences","studytime","freetime"]]
```

```python
X = st.drop("G3",axis=1)
Y = st["G3"]
X
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.16)
model = LinearRegression()
model.fit(x_train,y_train)

print("Accuracy on the training set: {:.3f}".format(model.score(x_train,y_train)))
print("Accuracy on the test set: {:.3f}".format(model.score(x_test,y_test)))

prediction = model.predict(x_test)

print(prediction)
```
