Importing the Dependencies

```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split,KFold,cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

Importing data file 

```python
st = pd.read_csv("student-mat.csv", sep=',')
```

```python
st.shape
```

```python
st.describe()
```

```python
st = st[["G1","G2","G3","age","Medu","Fedu","absences","studytime","freetime"]]
```

```python
X = st.drop("G3",axis=1)
y = st["G3"]
X
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.16, random_state=9)
model = LinearRegression()
model.fit(x_train,y_train)

prediction = model.predict(x_test)

print(prediction)

print("Accuracy on the training set: {:.3f}".format(model.score(x_train,y_train)))
print("Accuracy on the test set: {:.3f}".format(model.score(x_test,y_test)))


kfold = KFold(n_splits=5)

scores = cross_val_score(model,X,y,cv=kfold)

print("Cross Validation scores: ",scores)
print("Average CV Score: {:.3f}".format(scores.mean()))
```
